https://www.fbi.gov/how-we-can-help-you/need-an-fbi-service-or-more-information/ucr/publications#Hate-Crime%20Statistics (year 2013-2019) and https://crime-data-explorer.fr.cloud.gov/pages/downloads (year 2020).

```python
# drop Category rows, create a new Category and assign values
def assign_cat(df):
    big_bm = [0,9,17,23,26,29,32]
    cat = []
    for b in big_bm:
        cat.append(df['Bias_motivation'][b])

    df = df.drop(index=big_bm).reset_index(drop=True)
    df['Category'] = '0'

    bm = df['Bias_motivation']
    step_list = []
    cat_list = []
    start = 0
    for n in range (0,len(big_bm)-1):
        step = big_bm[n+1] - big_bm[n]
        step_list.append(step)
        end = start+step_list[n]-1
        list = bm.iloc[start:end].to_list()
        cat_list.append(list)
        start = end
        df['Category'] = np.where(bm.isin(cat_list[n]), cat[n], df['Category'])
        n+=1

    return df
```

```python
# convert categories to binary inputs
list = ['Race/Ethnicity/Ancestry','Religion','Sexual Orientation','Disability','Gender','Gender Identity']
def is_cat(df, category):
    cat = df[category]
    for i in range (0, len(list)):
        column_name = 'is_' + list[i]
        df[column_name] = np.where(cat.isin([list[i]]), 1, 0)
        i+=1
    return df
```

```python
# plot scatter plots, group by catrgories
def plot(df, h):  
    ax = sns.scatterplot(data=df, x='Year', y='Total_offenses', hue=h)
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    ax.set_xlim(2012,2020)
    plt.show()
```

```python
# import tables, data cleaning
years = np.linspace(2013, 2018, 6, dtype=int).tolist()
tables =[]
for y in years:
    n = y-2013
    table_path='./Data_%s/CSV_files_%s/Table_4_Offenses_Offense_Type_by_Bias_Motivation_%s.csv'%(y,y,y)
    tables.append(pd.read_csv(table_path, thousands=','))
    if y in range(2015, 2021):
        tables[n] = tables[n].drop(index=[1])

        # merge Anti-Arab to Anti-Other Race/Ethnicity/Ancestry
        oREA = tables[n]['Bias_motivation'] == 'Anti-Other Race/Ethnicity/Ancestry'
        arab = tables[n]['Bias_motivation'] == 'Anti-Arab'
        tables[n][oREA] = tables[n][oREA]+tables[n][arab].sum()
        tables[n]['Bias_motivation'][oREA] = 'Anti-Other Race/Ethnicity/Ancestry'
        tables[n] = tables[n].drop(index=tables[n][arab].index)

        # merge Anti-Other Religion
        r = tables[n].loc[19:25]
        oR = tables[n]['Bias_motivation'] == 'Anti-Other Religion'
        tables[n][oR] = tables[n][oR]+r.sum()
        tables[n]['Bias_motivation'][oR] = 'Anti-Other Religion'
        tables[n] = tables[n].drop(index=r.index)

    tables[n] = tables[n].drop(index=[0]).reset_index(drop=True)
    # add Year column
    tables[n]['Year'] = y
    tables[n] = assign_cat(tables[n])
     
# put all years data together
table_all = pd.concat(tables).reset_index(drop=True)
table_all
```

```python
# Bias motivation categories and their numbers of total offenses Group by Year
stacked = table_all[['Total_offenses','Year','Category']]
stacked = stacked.groupby(['Year','Category']).sum()
stacked
```

```python
# convert group by obj to list inputs for plots
def bar(list,i):
    b = []
    for n in range (0, len(list)):
        if (n-i)%6 == 0: b.append(list[n])
    return b
```

```python
# Stacked plot: an overview of our data
stacked = stacked.to_numpy().squeeze()
year = np.linspace(2013, 2018, 6, dtype=int).tolist()

bars = []
for n in range(0,6):bars.append(bar(stacked, n))

width = 0.35
fig, ax = plt.subplots()
l = ['Gender','Gender Identity', 'Sexual Orientation','Religion','Race/Ethnicity/Ancestry']
ax.bar(years, bars[0], width, label='Disability')

# swap 'Sexual Orientation' and 'Race/Ethnicity/Ancestry' to avoid overlapping
temp = bars[3]
bars[3] = bars[5]
bars[5] = temp

for i in range (1,6): 
    ax.bar(years, bars[i], width, bottom=bars[i-1],label=l[i-1])
ax.set_ylabel('Total_offenses')
ax.legend()
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

```python
# Plot the number of total offenses for each Bias Motivation category over year
for n in range(0,6):
    bars.append(bar(stacked, n))
    d = {'Year':year,'Total_offenses':bars[n]}
    df = pd.DataFrame(d)
    print(df.corr())
    ax = sns.lmplot(data=df, x='Year', y='Total_offenses').set(title=list[n])
    plt.show()
```

The regression plots above show the changes in the number of total offenses for each bias motivation category over 2013-2018. From the plot and the correlation matrix, we found that the strongest correlation is between year and the total number of offenses for Sexual Orientation bias, which is 0.909883. The weakest correlation is between year and the total number of offenses for Gender Identity bias, which is 0.208546. Higher correlation indicates a stronger linear relationship between the changes in number of total offenses (for the bias motivation category) and the changes in year. Therefore, we would focus on Sexual Orientation bias as the type of bias motivation in our research question and further analyze it with linear regression models.
